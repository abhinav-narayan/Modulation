```
# -*- coding: utf-8 -*-
"""
Binary ASK
"""
from numpy import arange,linspace,cos,pi,sin
import matplotlib.pyplot as plt
#import math
import random

# Start = 0
# Stop = 1
# limit = 4

# x = [random.randint(Start, Stop) for iter in range(limit)]

# print(x)


X=[]
x=[1,1,1,0,1,0]
for b in range(len(x)):
    if x[b] == 1:
        X.append(1)
    elif x[b] == 0:
        X.append(0)
print(X)
Y=[]

#n=linspace(0,len(X),num=len(X))
t=linspace(0,len(X),num=120)
phi=[]
for i in range (0,len(t)):
    phi.append(sin(2*pi*t[i]))

print(len(phi))
for ab in range(len(X)):
  if X[ab] == 1:
    for abc in range(20):
      Y.append(1)
  elif X[ab] == 0:
    for abcd in range(20):
      Y.append(0)
print(Y)
op = []
for i in range(len(t)):
  op.append(Y[i]*phi[i])

DM=[]
for k in range(0,len(op)):
  if op[k] >0 or op[k] < 0:
    DM.append(1)
  else:
    DM.append(0)
print(len(DM))

plt.plot(t,Y,drawstyle = 'steps-pre')            
plt.title('input')
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.show()

plt.plot(t,phi,'k')
plt.title('sinusoidal wave')
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.show()

plt.plot(t,op)
plt.title('ASK waveform')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()

plt.plot(t,DM,drawstyle = 'steps-pre')
plt.title('Demodulated signal')
plt.xlabel('Time information')
plt.ylabel('Amplitude')
plt.show()
```
